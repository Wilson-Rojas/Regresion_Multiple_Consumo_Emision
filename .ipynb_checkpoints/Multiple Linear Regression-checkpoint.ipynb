```python
Interpretación de los datos
FuelConsumption.csv:
Hemos descargado un conjunto de datos de consumo de combustible, FuelConsumption.csv, que contiene las clasificaciones de consumo de combustible específicas de cada modelo y las emisiones estimadas de dióxido de carbono para vehículos ligeros nuevos de venta al público en Canadá. Fuente del conjunto de datos:

AÑO DEL MODELO (p. ej., 2014)
MARCA (p. ej., Acura)
MODELO (p. ej., ILX)
CLASE DE VEHÍCULO (p. ej., SUV)
CILINDRADA (p. ej., 4.7)
CILINDROS (p. ej., 6)
TRANSMISIÓN (p. ej., A6)
TIPO DE COMBUSTIBLE (p. ej., z)
CONSUMO DE COMBUSTIBLE EN CIUDAD (L/100 km) (p. ej., 9.9)
CONSUMO DE COMBUSTIBLE EN CARRETERA (L/100 km) (p. ej., 8.9)
CONSUMO DE COMBUSTIBLE COMBINADO (L/100 km) (p. ej., 9.2
EMISIONES DE CO2 (g/km) p. ej. 182 --> bajo --> 0
```

```python
#Leemos todos los datos
df= pd.read_csv("FuelConsumptionCo2.csv")
# Darle forma 
df.head()
```

```python
#Seleccionemos algunas características que queremos utilizar para la regresión.
cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.head(9) 
```

```python
#Grafiquemos los valores de emisión con respecto al tamaño del motor
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS, color='blue')
plt.xlabel("Tamaño del Motor")
plt.ylabel("CO2 Emisiones")
plt.show()
```

```python
#### Creación de conjuntos de datos de entrenamiento y prueba
La división de entrenamiento/prueba implica dividir el conjunto de datos en conjuntos de entrenamiento y prueba, respectivamente, 
que son mutuamente excluyentes. Después, se entrena con el conjunto de entrenamiento y se prueba con el conjunto de prueba.
Esto proporcionará una evaluación más precisa de la precisión fuera de la muestra, ya que el conjunto de datos de prueba no forma 
parte del conjunto de datos utilizado para entrenar el modelo. Por lo tanto, nos permite comprender mejor la generalización de nuestro
modelo con nuevos datos.

Conocemos el resultado de cada punto de datos en el conjunto de datos de prueba, lo que lo convierte en una excelente opción para realizar pruebas.
Dado que estos datos no se han utilizado para entrenar el modelo, este desconoce el resultado de estos puntos. Por lo tanto, se trata, en esencia, 
de una prueba fuera de la muestra.

Dividamos nuestro conjunto de datos en conjuntos de entrenamiento y prueba. Aproximadamente el 80 % del conjunto de datos se utilizará para 
entrenamiento y el 20 % para pruebas. Creamos una máscara para seleccionar filas aleatorias mediante la función __np.random.rand()__:
```

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
#### Distribución de datos del tren

plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

```python
Multiple Regression Model
En realidad, existen múltiples variables que influyen en las emisiones de CO2. 
Cuando hay más de una variable independiente, el proceso se denomina regresión lineal múltiple.
Un ejemplo de regresión lineal múltiple es la predicción de las emisiones de CO2 mediante las 
características FUELCONSUMPTION_COMB, EngineSize y Cylinders de los automóviles. 
Lo bueno es que el modelo de regresión lineal múltiple es una extensión del modelo de regresión lineal simple.
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit (x, y)
# El coeficiente
print ('Coefficients: ', regr.coef_)
```

```python
Como se mencionó anteriormente, el coeficiente y la intersección son los parámetros de la línea ajustada.
Dado que se trata de un modelo de regresión lineal múltiple con tres parámetros, y que estos son la intersección y los coeficientes del hiperplano, 
sklearn puede estimarlos a partir de nuestros datos. Scikit-learn utiliza el método de Mínimos Cuadrados Ordinarios (MCO) para resolver este problema.

Mínimos Cuadrados Ordinarios (MCO)
MCO es un método para estimar los parámetros desconocidos en un modelo de regresión lineal. 
MCO selecciona los parámetros de una función lineal de un conjunto de variables explicativas 
minimizando la suma de los cuadrados de las diferencias entre la variable dependiente objetivo 
y las predichas por la función lineal. En otras palabras, intenta minimizar la suma de los errores 
cuadráticos (SSE) o el error cuadrático medio (MSE) entre la variable objetivo (y) y nuestro resultado predicho ( 𝑦̂ ) 
en todas las muestras del conjunto de datos.

MCO permite determinar los mejores parámetros mediante los siguientes métodos:

Resolución analítica de los parámetros del modelo mediante ecuaciones de forma cerrada
Uso de un algoritmo de optimización (descenso de gradiente, descenso de gradiente estocástico, método de Newton, etc.)
Predicción
```

```python
y_hat= regr.predict(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
x = np.asanyarray(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
y = np.asanyarray(test[['CO2EMISSIONS']])
print("Mean Squared Error (MSE) : %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x, y))
```

```python
Puntuación de regresión de varianza explicada:
Sea 𝑦̂ el resultado objetivo estimado, y el resultado objetivo correspondiente (correcto) 
y Var la varianza (el cuadrado de la desviación estándar). La varianza explicada se estima de la siguiente manera:

𝚎𝚡𝚙𝚕𝚊𝚒𝚗𝚎𝚍𝚅𝚊𝚛𝚒𝚊𝚗𝚌𝚎(𝑦,𝑦̂ )=1−𝑉𝑎𝑟{𝑦−𝑦̂ }𝑉𝑎𝑟{𝑦}

La mejor puntuación posible es 1.0; los valores más bajos son peores.
```

```python
regr_2= linear_model.LinearRegression()
x = np.asanyarray(train[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY']])
y = np.asanyarray(train[['CO2EMISSIONS']])
regr_2.fit (x, y)
print ('Coefficients: ', regr.coef_)
y_= regr_2.predict(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY']])
x = np.asanyarray(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY']])
y = np.asanyarray(test[['CO2EMISSIONS']])
print("Mean Squared Error (MSE) : %.2f"
      % np.mean((y_ - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr_2.score(x, y))
```
